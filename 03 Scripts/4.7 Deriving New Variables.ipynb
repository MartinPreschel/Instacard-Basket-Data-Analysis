## If-Statements with User-Defined Functions

```python
# Define the price_label

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function 

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check the values in the new column

df['price_range'].value_counts(dropna = False)
```

```python
# Why not check what the most expensive product within the subset is?

df['prices'].max()
```

## If-Statements with For-Loops

```python
#To start, you need to know on which day most orders take place

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create a new column, âbusiest day,â that will contain one of three different values: âBusiest day,â âLeast busy,â and âRegularly busy.

result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Create column busiest_days from "orders_day_of_week" to identify two busiest days (0, 1), slowest days (4, 3), 
# or regularly busy (all other days)

result_2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest day")
  elif value == 3 or value == 4:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
# Create new 'busiest_days' column in df_ords_prods_merge to view results in context

df_ords_prods_merge['Busiest days'] = result_2
```

## Task 4)

```python
#To start, I need to know on which hour most orders take place by making a order frequency analysis

df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating busiest_period_of_day column to categorize order times

result_3 = []

# Defining categories based on order frequency analysis

most_orders = [10, 11, 12, 13, 14, 15, 16] # Peak order hours
average_orders = [7, 8, 9, 17, 18, 19, 20, 21, 22, 23] # Moderate order hours
fewer_orders = [0, 1, 2, 3, 4, 5, 6] # Low activity hours

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in most_orders:
    result_3.append("Most orders")
  elif value in average_orders:
    result_3.append("Average orders")
  else:
    result_3.append("Fewest orders")
```

```python
# Creating the new column in the DataFrame

df_ords_prods_merge["busiest_period_of_day"] = result_3
```
